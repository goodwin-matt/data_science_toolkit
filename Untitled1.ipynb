```python
filename = 'sample.csv'
n=10
```

```python
sample = pd.read_csv(filename)
num_rows = sample.shape[0]
sample['Typical Price'] = (sample['High'] + sample['Low'] + sample['Close'])/3.0
sample['Money Flow'] = sample['Typical Price']*sample['Volume']
sample['NegPos'] = 0
sample.loc[1:,'NegPos'] = sample.loc[1:,'Typical Price'].values  - sample.loc[:(num_rows-2),'Typical Price'].values
sample['Positive Money Flow'] = np.nan
sample['Negative Money Flow'] = np.nan
sample.loc[(sample.NegPos > 0),'Positive Money Flow'] = sample.loc[(sample.NegPos > 0),'Money Flow']
sample.loc[(sample.NegPos < 0),'Negative Money Flow'] = sample.loc[(sample.NegPos < 0),'Money Flow']
sample['Positive Money Flow Sum'] = np.nan
sample['Negative Money Flow Sum'] = np.nan
rolling_sum_positive = sample['Positive Money Flow'].rolling(n, min_periods=1).sum()
sample.loc[10:,'Positive Money Flow Sum'] = rolling_sum_positive[n:,]
rolling_sum_negative = sample['Negative Money Flow'].rolling(n, min_periods=1).sum()
sample.loc[10:,'Negative Money Flow Sum'] = rolling_sum_negative[n:,]
sample['Money Ratio'] = sample['Positive Money Flow Sum']/sample['Negative Money Flow Sum']
sample['Money Flow Index'] = 100*sample['Money Ratio']/(1 + sample['Money Ratio'])
```

```python
sample.loc[:,['Day','Open','High','Low','Close','Volume','Typical Price','Positive Money Flow','Negative Money Flow','Positive Money Flow Sum','Negative Money Flow Sum','Money Flow Index']]
```
